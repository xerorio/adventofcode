```python
# PART 1

import pyperclip

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

import re

def read_file():
    with open('13.txt') as f:
        return [line.split('\n') for line in f.read().split('\n\n')]

packet_pairs = read_file()

def interpret_list_structure(data):
    """
    Parse and interpret nested list structures from string representation.
    """
    import ast
    try:
        # Safely evaluate the structure
        return ast.literal_eval(data)
    except Exception as e:
        raise ValueError(f"Failed to interpret list structure: {e}")

def compare_pairs(left, right):
    print(left)
    print(right)
    print('intrepeting...')

    left_parsed = interpret_list_structure(left)
    right_parsed = interpret_list_structure(right)
    print(left_parsed)
    print(right_parsed)

    print('')

    left_first = left_parsed[0]
    right_first = right_parsed[0]

    if type(left_first) == list:
        compare_pairs(left_first, right_first)

    return True

valid_indices = 0
for i, pair in enumerate(packet_pairs):
    l = pair[0]
    r = pair[1]

    if compare_pairs(l, r):
        valid_indices += i

print(f'ANSWER: {print_and_copy(valid_indices)}')
```
