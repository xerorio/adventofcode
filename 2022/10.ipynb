```python
# PART 1

def read_file():
    with open('10.txt') as f:
        return [line.split(' ') for line in f.read().splitlines()]

operations = read_file()

# values found in the cycle represent the value of X *DURING* the cycle
cycles = [1]

for operation in operations:
    if len(operation) == 1: # noop
        cycles.append(cycles[-1])
    else:
        next_num = int(operation[1])
        cycles.append(cycles[-1]) # same as skipping a cycle
        cycles.append(cycles[-1] + next_num) # then this is the value during the next cycle

# print(cycles)

total = 0

# important cycles: 20th, 60th, 100th, 140th, 180th, and 220th
for n in range(20, 221, 40):
    print(f'{n} * {cycles[n-1]} = {n * cycles[n-1]}')
    total += (n * cycles[n-1])

print(f'ANSWER: {total}')
```

```python
# PART 2

def read_file():
    with open('10.txt') as f:
        return [line.split(' ') for line in f.read().splitlines()]

operations = read_file()

# position along the CRT line as an index
sprite_position = 1

CRT = ['']
current_row = 0

def add_to_CRT():
    global sprite_position
    global CRT
    global current_row

    if sprite_position > 40:
        sprite_position -= 40

    length = len(CRT[current_row])

    if length > 39:
        current_row += 1
        length -= 40
        CRT.append('')

    # character needs to be the sprite position instead
    if sprite_position - 1 <= length <= sprite_position + 1:
        char = '#'
    else:
        char = '.'

    CRT[current_row] += char

for operation in operations:
    cycle_num = sum([len(line) for line in CRT]) + 1
    if len(operation) == 1: # noop
        print(f'Start cycle   {cycle_num}: begin executing noop')
        print(f'During cycle  {cycle_num}: CRT draws pixel in position {cycle_num - 1}')
        add_to_CRT()
    else: # addx
        print(f'Start cycle   {cycle_num}: begin executing addx {operation[1]}')
        print(f'During cycle  {cycle_num}: CRT draws pixel in position {cycle_num - 1}')
        addx = int(operation[1])
        add_to_CRT()
        print(f'Current CRT row: {CRT[-1]}')
        print('')

        cycle_num += 1

        print(f'During cycle  {cycle_num}: CRT draws pixel in position {cycle_num - 1}')
        add_to_CRT()
        sprite_position += addx % 40
        print(f'Current CRT row: {CRT[-1]}')
        print(f'End of cycle  {cycle_num}: finish executing addx {operation[1]} (Register X is now {sprite_position})')
    
    print('')
    
    if len(CRT) > 6: # stop if cycles go past 240
        break

for line in CRT:
    # print(line)
    print(line.replace('.', '  ').replace('#', '##'))

# output should look like: REHPRLUB
```
