```python
import pyperclip

def copy_and_return(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('01.txt') as f:
        return f.read().splitlines()

data = read_file()
```

```python
# PART 1

dial_pointer = 50

ans = 0

for line in data:
    direction = line[0]
    rotation_amount = int(line[1:])

    if rotation_amount > 99:
        rotation_amount = int(str(rotation_amount)[-2:])

    if direction == "R":
        dial_pointer += rotation_amount
    else:
        dial_pointer -= rotation_amount
    
    """
    if the dial pointer is negative, do 100 minus dial pointer until it is between 0 and 99
    if the dial pointer is more than 99, take away 100 until it is between 0 and 99
    """
    if dial_pointer < 0:
        while not(0 <= dial_pointer <= 99):
            dial_pointer = 100 + dial_pointer
    elif dial_pointer > 99:
        while not(0 <= dial_pointer <= 99):
            dial_pointer -= 100

    if dial_pointer == 0:
        ans += 1
    
    print(f'{line}\tcurrent point: {dial_pointer}\tcounter: {ans}')

print(f'ANSWER: {copy_and_return(ans)}')
```

```python
# PART 2

dial_pointer = 50

ans = 0

for line in data:
    direction = line[0]
    rotation_amount = int(line[1:])

    # for each 100 in a rotation, add 1 to the times passing 0
    # rotation_amount becomes the remainder of dividing by 100
    num_times_passing_zero, rotation_amount = divmod(rotation_amount, 100)

    ans += num_times_passing_zero

    if rotation_amount == 0:
        continue
    
    passed_zero = False

    """
    if rotation_amount is going to the right and (100 - rotation_amount):
    - is equal to dial_pointer, it will stop at 0
    - is more than dial_pointer, it will pass 0
    - is less than dial_pointer, all good

    if rotation_amount is going to the left and rotation_amount:
    - is equal to dial_pointer, it will stop at 0
    - is more than dial_pointer, it will pass 0
    - is less than dial_pointer, all good
    """
    if direction == "R":

        if dial_pointer == (100 - rotation_amount):
            dial_pointer = 0
        elif dial_pointer > (100 - rotation_amount):
            dial_pointer += (rotation_amount - 100)
            ans += 1
            passed_zero = True
        elif dial_pointer < (100 - rotation_amount):
            dial_pointer += rotation_amount

    else:
        
        if dial_pointer == rotation_amount:
            dial_pointer = 0
        elif dial_pointer > rotation_amount:
            dial_pointer -= rotation_amount
        elif dial_pointer < rotation_amount:
            dial_pointer += (100 - rotation_amount)
            ans += 1
            passed_zero = True
    
    if passed_zero:
        print(f'{line}\tcurrent point: {dial_pointer}\tcounter: {ans} (passed zero)')
    else:
        print(f'{line}\tcurrent point: {dial_pointer}\tcounter: {ans}')

print(f'ANSWER: {copy_and_return(ans)}')
```
