```python
# PART 1

import pyperclip

def copy_and_return(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('02.txt') as f:
        return f.read()

data = [x.split('-') for x in read_file().split(',')]

invalid_ids_total = 0

def check_if_num_is_invalid(x):
    s = str(x)
    for a in range(len(s)):
        for b in range(1, a):
            pattern = s[a:b]
            # pattern = r'(.{' + str(length) + r'})\\1+'
            if pattern in s:
                return True
    return False

for data_range in data:
    for n in range(int(data_range[0]), int(data_range[1]) + 1):
        if check_if_num_is_invalid(n):
            invalid_ids_total += n

print(f'ANSWER: {copy_and_return(invalid_ids_total)}')
```
