```python
# PART 1

import pyperclip

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('15.txt') as f:
        return f.read().split('\n\n')

grid, moves = read_file()

grid = [list(line) for line in grid.split('\n')]
n = len(grid)
m = len(grid[0])

moves = moves.replace('\n', '')

for i in range(n):
    for j in range(m):
        if grid[i][j] == '@':
            x, y = (i, j)
            break

dirs = {
    '<': (0, -1),
    '>': (0, 1),
    '^': (-1, 0),
    'v': (1, 0)
}

def in_grid(i, j):
    return (0 <= i < n) and (0 <= j < m)

def move_robot(dir):
    global x, y, grid

    next_i, next_j = x + dir[0], y + dir[1]
    if not in_grid(next_i, next_j):
        return
    
    di, dj = x, y
    while in_grid(di, dj):
        di += dir[0]
        dj += dir[1]

        if not in_grid(di, dj):
            break
        
        if grid[di][dj] == '#':
            break

        if grid[di][dj] == '.':
            grid[di][dj] = 'O'
            grid[x][y] = '.'

            x, y = x + dir[0], y + dir[1]
            grid[x][y] = '@'
            break

for move in moves:
    move_robot(dirs[move])

ans = 0
for i in range(n):
    for j in range(m):
        if grid[i][j] == 'O':
            ans += (100 * i) + j

print(f'ANSWER: {print_and_copy(ans)}')
```

```python
# PART 2

import pyperclip

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

from tqdm import tqdm

def read_file():
    with open('15.txt') as f:
        return f.read().split('\n\n')

grid, moves = read_file()

moves = moves.replace('\n', '')

grid = [list(line) for line in grid.split('\n')]
n = len(grid)

boxes = []
walls = []
for i in range(n):
    for j in range(n):
        if grid[i][j] == '@':
            x, y = i, j*2
        elif grid[i][j] == 'O':
            boxes.append([i, j*2])
        elif grid[i][j] == '#':
            walls.append([i, j*2])
            walls.append([i, j*2+1])
    
def in_grid(i, j):
    return (0 <= i < n) and (0 <= j < 2*n)

def move_robot(dir):
    global x, y, grid, boxes, walls

    next_i, next_j = x + dir[0], y + dir[1]
    if not in_grid(next_i, next_j):
        return
    
    if [next_i, next_j] in walls:
        return

    stack = []
    if [next_i, next_j] in boxes:
        stack.append([next_i, next_j])
    if [next_i, next_j-1] in boxes:
        stack.append([next_i, next_j-1])

    can_move = True

    seen = set()
    while len(stack) > 0:
        top_i, top_j = stack.pop()
        ni, nj = top_i + dir[0], top_j + dir[1]

        if not in_grid(ni, nj):
            can_move = False
            break
            
        if [ni, nj] in walls or [ni, nj+1] in walls:
            can_move = False
            break

        if (top_i, top_j) in seen:
            continue
        seen.add((top_i, top_j))

        if [ni, nj] in boxes:
            stack.append([ni, nj])
        if [ni, nj-1] in boxes:
            stack.append([ni, nj-1])
        if [ni, nj+1] in boxes:
            stack.append([ni, nj+1])
    
    if not can_move:
        return
    
    for i, box in enumerate(boxes):
        if tuple(box) in seen:
            boxes[i][0] += dir[0]
            boxes[i][1] += dir[1]
    
    x += dir[0]
    y += dir[1]

dirs = {
    '<': [0, -1],
    'v': [1, 0],
    '>': [0, 1],
    '^': [-1, 0]
}

def print_grid(boxes, walls, x, y):
    for i in range(n):
        for j in range(n*2):
            if [i, j] in walls:
                print('#', end='')
            elif [i, j] in boxes:
                print('[', end='')
            elif [i, j-1] in boxes:
                print(']', end='')
            elif (i, j) == (x, y):
                print('@', end='')
            else:
                print('.', end='')
        print()

# print_grid(boxes, walls, x, y)

for move in tqdm(moves):
    move_robot(dirs[move])

# print_grid(boxes, walls, x, y)

ans = 0
for i, j in boxes:
    ans += i*100 + j

print(f'ANSWER: {print_and_copy(ans)}')
```
