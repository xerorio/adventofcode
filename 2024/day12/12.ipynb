```python
# PART 2

import pyperclip
from collections import defaultdict

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('12.txt') as f:
        return [list(line) for line in f.read().splitlines()]

grid = read_file()
n, m = len(grid), len(grid[0])

# to find regions, search recursively until you find edges
dd = [(-1, 0), (1, 0), (0, 1), (0, -1)]

labelled_locations = {}
# this dfs creates a dictionary where the key is the location and
# the value is the region that it's found in (represented by an id)
def dfs(x, y, plant_type, id):
    if x in range(n) and y in range(m):
        if (x, y) in labelled_locations:
            return
        if grid[x][y] == plant_type:
            labelled_locations[x, y] = id
            for dx, dy in dd:
                dfs(x + dx, y + dy, plant_type, id)

region_id = 0
for i in range(n):
    for j in range(m):
        if (i, j) not in labelled_locations:
            dfs(i, j, grid[i][j], region_id)
            region_id += 1

# converts the "labelled_locations" dictionary into another dictionary where
# the key is the "region_id" and the value is a set of the locations of that region
region_locations = defaultdict(set)
for k, v in labelled_locations.items():
    region_locations[v].add(k)

total_price = 0
for locations in region_locations.values():
    area = len(locations)
    perimeter_locations_next_to_each_other = []
    
    for i, j in locations:
        for dx, dy in dd:
            ii = i + dx
            jj = j + dy
            if ii not in range(n) or jj not in range(m) or (ii, jj) not in locations:
                perimeter_locations_next_to_each_other.append(((i, j), (ii, jj)))
    
    perimeter_locations_next_to_each_other = set(perimeter_locations_next_to_each_other)
    sides = set()
    for p1, p2 in perimeter_locations_next_to_each_other:
        keep = True
        for dx, dy in [(1, 0), (0, 1)]:
            p1n = (p1[0] + dx, p1[1] + dy)
            p2n = (p2[0] + dx, p2[1] + dy)
            if (p1n, p2n) in perimeter_locations_next_to_each_other:
                keep = False
        if keep:
            sides.add((p1, p2))
    
    total_price += area * len(sides)

print(f'ANSWER: {print_and_copy(total_price)}')
```
