```python
# PART 1

from itertools import product

def read_file():
    with open('07.txt') as f:
        return [line.split(': ') for line in f.read().splitlines()]

equations = read_file()

total = 0

for i, equation in enumerate(equations):
    target_value = int(equation[0])
    nums = list(map(int, equation[1].split(' ')))

    def test(combo):
        ans = nums[0]
        for i in range(1, len(nums)):
            if combo[i-1] == '+':
                ans += nums[i]
            else:
                ans *= nums[i]
        return ans

    for combo in product('*+', repeat=len(nums)-1):
        if test(combo) == target_value:
            total += target_value
            print(f'[{i:02}/{len(equations)}]', combo, target_value)
            break

print(f'ANSWER: {total}')
```

```python
# PART 2

from datetime import datetime

start = datetime.now()

from itertools import product

def read_file():
    with open('07.txt') as f:
        return [line.split(': ') for line in f.read().splitlines()]

equations = read_file()

total = 0

for i, equation in enumerate(equations):
    target_value = int(equation[0])
    nums = list(map(int, equation[1].split(' ')))

    def test(combo):
        ans = nums[0]
        for i in range(1, len(nums)):
            if combo[i-1] == '|':
                ans = int(f'{ans}{nums[i]}')
            elif combo[i-1] == '+':
                ans += nums[i]
            else:
                ans *= nums[i]
        return ans

    for combo in product('*+|', repeat=len(nums)-1):
        if test(combo) == target_value:
            total += target_value
            print(f'[{i:02}/{len(equations)}]', combo, target_value)
            break

print(f'Time taken: {datetime.now() - start}')
print(f'ANSWER: {total}')
```
