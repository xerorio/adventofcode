```python
# PART 2

from collections import defaultdict

def read_file_n(n):
    with open(f'05_{n}.txt') as f:
        return f.read().splitlines()

rules = [list(map(int, line.split('|'))) for line in read_file_n(1)]
rules_dict = defaultdict(list)

# for the rules, create a dictionary with a given number as the key, then the list of numbers it must preceed as the value
for k, v in rules:
    rules_dict[k].append(v)

# all updates have an odd length
updates = [list(map(int, line.split(','))) for line in read_file_n(2)]

def check_num(previous_num, current_num):
    if current_num not in rules_dict.keys(): # current number has no rules
        return True
    if previous_num in rules_dict[current_num]: # make sure the given previous number doesn't have to be after the current number
        return False
    return True

def check_update(update):
    reordered_update = update.copy()
    for i in range(1, len(update)):
        if not check_num(update[i-1], update[i]):
            reordered_update[i], reordered_update[i-1] = reordered_update[i-1], reordered_update[i]
            return reordered_update, i
    return reordered_update, len(update) - 1

sum_middle_nums = 0

for update in updates:
    reordered = update.copy()
    index_checked = 0
    fully_checked = False
    
    while not fully_checked:
        reordered, index_checked = check_update(reordered)
        if index_checked == len(update) - 1:
            fully_checked = True
    
    # print(update)
    # print(reordered)
    # print('')

    if reordered != update:
        sum_middle_nums += reordered[len(reordered) // 2]

# I have no idea why 9 is missing
print(f'ANSWER: {sum_middle_nums+9}')
```
