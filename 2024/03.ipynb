```python
# PART 1

import re

def read_file():
    with open('03.txt') as f:
        return ''.join(f.readlines())

input_text = read_file()

regex_str = r'mul\((\d{1,3}),(\d{1,3})\)'

matches = re.findall(regex_str, input_text)
total = 0

for match in matches:
    total += int(match[0]) * int(match[1])

print(f'ANSWER: {total}')
```

```python
# PART 2

def read_file():
    with open('03.txt') as f:
        return ''.join(f.readlines())

input_text = read_file()

regex_str = r"mul\((\d{1,3}),(\d{1,3})\)|(do\(\)|don't\(\))"

matches = re.findall(regex_str, input_text)

total = 0
enabled = True

for match in matches:
    if match[2] == '' and enabled:
        total += int(match[0]) * int(match[1])
    else:
        if match[2] == 'do()':
            enabled = True
        else:
            enabled = False

print(f'ANSWER: {total}')
```
