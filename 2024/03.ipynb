```python
# PART 1

import re

def read_file():
    with open('03.txt') as f:
        return ''.join(f.readlines())

input_text = read_file()

regex_str = r'(mul\(\d{1,3},\d{1,3}\))'

matches = re.findall(regex_str, input_text)

def multiply_numbers(instruction):
    nums_to_multiply = tuple(map(int, instruction.replace('mul(', '').replace(')', '').split(',')))
    return nums_to_multiply[0] * nums_to_multiply[1]

total = 0

for match in matches:
    total += multiply_numbers(match)

print(f'ANSWER: {total}')
```

```python
# PART 2

def read_file():
    with open('03.txt') as f:
        return ''.join(f.readlines())

input_text = read_file()

regex_str = r"(mul\(\d{1,3},\d{1,3}\))|(do\(\))|(don't\(\))"
# regex_str = r"(mul\(\d{1,3},\d{1,3}\))|(do\(\))"

def clean_up_regex_output(m):
    return ''.join(m)

matches = re.findall(regex_str, input_text)
clean_matches = list(map(clean_up_regex_output, matches))

def multiply_numbers(instruction):
    nums_to_multiply = tuple(map(int, instruction.replace('mul(', '').replace(')', '').split(',')))
    return nums_to_multiply[0] * nums_to_multiply[1]

total = 0
enabled = True

for match in clean_matches:
    if match == "don't()":
        enabled = False
        continue
    if match == "do()":
        enabled = True
        continue
    
    if enabled:
        total += multiply_numbers(match)

print(f'ANSWER: {total}')
```
