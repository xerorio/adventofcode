```python
# PART 1

import pyperclip

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('17.txt') as f:
        a, b = f.read().split('\n\n')
        a = {f'{x.split(' ')[1][0]}': int(x.split(' ')[-1]) for x in a.split('\n')}
        b = list(map(int, b.split(' ')[1].split(',')))
        return a, b

registers, program = read_file()

def get_combo_operand(oper):
    if 0 <= oper <= 3: return oper
    if oper == 4: return registers['A']
    if oper == 5: return registers['B']
    if oper == 6: return registers['C']
    raise RuntimeError(f'Invalid operand: {oper}')

instruction_pointer = 0
output = []

while instruction_pointer < len(program):
    op_code = program[instruction_pointer]
    input_operand = program[instruction_pointer + 1]

    if op_code == 0:
        combo_operand = get_combo_operand(input_operand)
        registers['A'] = registers['A'] >> combo_operand
    elif op_code == 1:
        registers['B'] = registers['B'] ^ input_operand
    elif op_code == 2:
        combo_operand = get_combo_operand(input_operand)
        registers['B'] = int(combo_operand % 8)
    elif op_code == 3:
        if registers['A'] != 0:
            instruction_pointer = input_operand
            continue
    elif op_code == 4:
        registers['B'] = registers['B'] ^ registers['C']
    elif op_code == 5:
        combo_operand = get_combo_operand(input_operand)
        output.append(combo_operand % 8)
    elif op_code == 6:
        combo_operand = get_combo_operand(input_operand)
        registers['B'] = registers['A'] >> combo_operand
    elif op_code == 7:
        combo_operand = get_combo_operand(input_operand)
        registers['C'] = registers['A'] >> combo_operand
    
    instruction_pointer += 2

print(f'ANSWER: {print_and_copy(','.join(map(str, output)))}')
```

```python
# PART 2

import pyperclip

def print_and_copy(x):
    pyperclip.copy(x)
    return(x)

def read_file():
    with open('17.txt') as f:
        a, b = f.read().split('\n\n')
        # a = {f'{x.split(' ')[1][0]}': int(x.split(' ')[-1]) for x in a.split('\n')}
        b = list(map(int, b.split(' ')[1].split(',')))
        return b

program = read_file()
print(program)

assert program[-2:] == [3, 0], 'program does not end with JNZ 0'

def find(target, ans):
    if target == []: return ans
    for t in range(8):
        a = ans << 3 | t
        b = 0
        c = 0
        output = None
        adv3 = False

        def get_combo_operand(oper):
            if 0 <= oper <= 3: return oper
            if oper == 4: return a
            if oper == 5: return b
            if oper == 6: return c
            raise RuntimeError(f'Invalid operand: {oper}')

        for pointer in range(0, len(program) - 2, 2):
            ins = program[pointer]
            operand = program[pointer + 1]

            if ins == 0:
                assert not adv3, 'program has multiple ADVs'
                assert operand == 3, 'program has ADV with operand other than 3'
                adv3 == True
            elif ins == 1:
                b = b ^ operand
            elif ins == 2:
                b = get_combo_operand(operand) % 8
            elif ins == 3:
                raise AssertionError('program has JNZ inside expected loop body')
            elif ins == 4:
                b = b ^ c
            elif ins == 5:
                assert output is None, 'program has multiple OUT'
                output = get_combo_operand(operand) % 8
            elif ins == 6:
                b = a >> get_combo_operand(operand)
            elif ins == 7:
                c = a >> get_combo_operand(operand)
            if output == target[-1]:
                sub = find(target[:-1], a)
                if sub is None: continue
                return sub

print(f'ANSWER: {print_and_copy(find(program, 0))}')
```
